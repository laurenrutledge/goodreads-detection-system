```python
import pandas as pd

df = pd.read_csv("datasets/processed/goodreads_reviews_with_nlp_features_substantiveness.csv")

# Overview
print(df.head())          # first 5 rows
print(df.tail())          # last 5 rows
print(df.info())          # column types, non-null counts
print(df.shape)           # (rows, columns)
```

```python
print(df.describe())          # numeric columns: mean, std, min, max, quartiles
print(df.describe(include='all'))  # include object/categorical columns too
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for a numeric feature
df['word_count'].hist(bins=30)
plt.show()

# Boxplot to detect outliers
sns.boxplot(x=df['word_count'])
plt.show()

# Pairwise plots of a few columns
sns.pairplot(df[['n_votes', 'sentence_count','word_count','avg_words_per_sentence','lexical_diversity']])
plt.show()
```
